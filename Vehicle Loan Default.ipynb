```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the data
data = pd.read_csv('Downloads/vehicle loan default.csv')
data
```

```python
# checking the columns
data.columns
```

```python
# dropping unnecessary columns
data.drop(['BRANCH_ID','DATE_OF_BIRTH','MANUFACTURER_ID','SUPPLIER_ID','EMPLOYEE_CODE_ID','STATE_ID','PERFORM_CNS_SCORE',
     'PERFORM_CNS_SCORE_DESCRIPTION','AVERAGE_ACCT_AGE','CREDIT_HISTORY_LENGTH','DISBURSAL_DATE'],axis = 1, inplace = True)
```

# Data Summary

```python
data.describe()
```

```python
# checking for null values
data.isna().sum()
```

```python
# dropping empty cells
data.dropna(inplace=True)
```

```python
data.info()
```

# Correlation

```python
correlation = data.corr()
correlation['LOAN_DEFAULT'].reset_index()
```

# Pre-Processing

```python
# labelling the string data
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
data['EMPLOYMENT_TYPE'] = label_encoder.fit_transform(data['EMPLOYMENT_TYPE'])
```

```python
data
```

```python
# separating target and feature variables
x = data.drop('LOAN_DEFAULT',axis=1)
y = data['LOAN_DEFAULT']
```

```python
# splitting the data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train),len(x_test),len(y_train),len(y_test)
```

```python
# importing libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
```

# Logistic Regression

```python
log = LogisticRegression(max_iter=500)
log_reg = log.fit(x_train,y_train)
```

```python
log_reg.predict(x_test)
```

```python
# train_score
log_train_score = log_reg.score(x_train,log_reg.predict(x_train))
log_train_score
```

```python
# test_score
log_test_score = log_reg.score(x_test,log_reg.predict(x_test))
log_test_score
```

# Decision Tree

```python
tree = DecisionTreeClassifier().fit(x_train,y_train)
```

```python
# train_score
tree_train_score = tree.score(x_train,tree.predict(x_train))
tree_train_score
```

```python
# test_score
tree_test_score = tree.score(x_test,tree.predict(x_test))
tree_test_score
```

# Random Forest  Classifier

```python
rf = RandomForestClassifier().fit(x_train,y_train)
```

```python
# train_score
rf_train_score = rf.score(x_train,rf.predict(x_train))
rf_train_score
```

```python
# test_score
rf_test_score = rf.score(x_test,rf.predict(x_test))
rf_test_score
```

# K Neighbors Classifier

```python
knn = KNeighborsClassifier().fit(x_train,y_train)
```

```python
# train_score
knn_train_score = knn.score(x_train,knn.predict(x_train))
knn_train_score
```

```python
# test_score
knn_test_score = knn.score(x_test,knn.predict(x_test))
knn_test_score
```

# XGBoost

```python
xgb = XGBClassifier().fit(x_train,y_train)
```

```python
# train_score
xgb_train_score = xgb.score(x_train,xgb.predict(x_train))
xgb_train_score
```

```python
#  test_score
xgb_test_score = xgb.score(x_test,xgb.predict(x_test))
xgb_test_score
```

# comparing models

```python
models = pd.DataFrame({
    'model': ['log_reg','tree','rf','knn','xgb'],
    'train_score': [log_train_score,tree_train_score,rf_train_score,knn_train_score,xgb_train_score],
    'test_score': [log_test_score,tree_test_score,rf_test_score,knn_test_score,xgb_test_score]
})
```

```python
models.sort_values("test_score",ascending=False)
```
